```python
import glob
from zipfile import ZipFile

for file in glob.glob("./input/**/*.zip", recursive=True):
    print(f"Extracting {file}")

    # Remove zip extension
    trimmed_path = file[:-4]

    with ZipFile(file, 'r') as zipFile: 
        zipFile.extractall( path=trimmed_path) 
```
